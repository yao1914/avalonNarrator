```python
import pyttsx3
import time

class AdvancedAvalonNarrator:
    def __init__(self):
        self.engine = pyttsx3.init()
        self.engine.setProperty('rate', 160)
        
        # Official Avalon distribution: { total_players: (good_count, evil_count) }
        self.player_limits = {
            5: (3, 2),
            6: (4, 2),
            7: (4, 3),
            8: (5, 3),
            9: (6, 3),
            10: (6, 4)
        }

    def speak(self, text, pause=2.5):
        print(f"[Narrator]: {text}")
        self.engine.say(text)
        self.engine.runAndWait()
        time.sleep(pause)

    def setup_game(self):
        print("=== AVALON NARRATOR SETUP ===")
        
        # 1. Get player count
        while True:
            try:
                players = int(input("Enter number of players (5-10): "))
                if players in self.player_limits:
                    break
                print("Invalid number. Avalon is played with 5 to 10 players.")
            except ValueError:
                print("Please enter a valid number.")

        max_good, max_evil = self.player_limits[players]
        print(f"\nFor {players} players, you need: {max_good} Good and {max_evil} Evil.")
        print("-" * 30)
        print("Merlin (Good) and Assassin (Evil) are automatically included.")

        # 2. Select Optional Roles
        # Format: (Role Name, Side)
        optional_roles = {
            "Percival": "Good",
            "Morgana": "Evil",
            "Mordred": "Evil",
            "Oberon": "Evil"
        }
        
        active_roles = {"Merlin": True, "Assassin": True}
        current_good = 1 # Merlin
        current_evil = 1 # Assassin

        for role, side in optional_roles.items():
            # Check if we even have room left for this side
            if side == "Good" and current_good >= max_good:
                continue
            if side == "Evil" and current_evil >= max_evil:
                continue
                
            choice = input(f"Include {role} ({side})? (y/n): ").strip().lower()
            if choice == 'y':
                active_roles[role] = True
                if side == "Good":
                    current_good += 1
                else:
                    current_evil += 1
            else:
                active_roles[role] = False

        # Fill the rest with generic vanilla roles
        print(f"\nSetup Complete!")
        print(f"Good roles: {current_good} Special + {max_good - current_good} Vanilla Loyal Servants")
        print(f"Evil roles: {current_evil} Special + {max_evil - current_evil} Vanilla Minions")
        
        input("\nPress Enter when everyone has their cards and is ready to begin...")
        self.run_narration(active_roles)

    def run_narration(self, roles):
        print("\n--- NARRATION STARTING ---")
        self.speak("Everyone, close your eyes and extend your hand into a fist in front of you.", 3)

        # 1. Minions acknowledge each other (Oberon stays asleep, Mordred wakes up)
        if roles.get("Oberon"):
            self.speak("All minions of Mordred, except Oberon, open your eyes and look around to acknowledge each other.")
        else:
            self.speak("Minions of Mordred, open your eyes and look around to acknowledge each other.")
        
        self.speak("Minions of Mordred, close your eyes.")

        # 2. Merlin sees Minions (Mordred stays hidden)
        if roles.get("Mordred"):
            self.speak("Minions of Mordred, except Mordred himself, extend your thumbs so that Merlin may know you.")
        else:
            self.speak("Minions of Mordred, extend your thumbs so that Merlin may know you.")
            
        self.speak("Merlin, open your eyes and see the minions of Mordred.")
        self.speak("Merlin, close your eyes. Minions, put your thumbs down.")

        # 3. Percival sees Merlin & Morgana
        if roles.get("Percival"):
            if roles.get("Morgana"):
                self.speak("Merlin and Morgana, extend your thumbs so that Percival may know you.")
            else:
                self.speak("Merlin, extend your thumb so that Percival may know you.")
                
            self.speak("Percival, open your eyes and see the thumbs.")
            self.speak("Percival, close your eyes. Thumbs down.")

        # 4. Wake up
        self.speak("Everyone, open your eyes.", pause=0)
        print("--- NARRATION COMPLETE ---")

if __name__ == "__main__":
    narrator = AdvancedAvalonNarrator()
    narrator.setup_game()
```
